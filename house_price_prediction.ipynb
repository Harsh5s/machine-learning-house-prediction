```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from utils import *
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
%matplotlib inline


```

Data Normalization

```python
df = df.iloc[:,1:]
df_norm = (df - df.mean()) / df.std()
df_norm.head()
```

Select X Labels

```python
X = df_norm.iloc[:, :5]
X.head()
```

Select Y Labels

```python
Y = df_norm.iloc[:, -1]
Y.head()
```

Feature And Label Values

```python
X_arr = X.values
Y_arr = Y.values
```

```python
X_arr
```

```python
Y_arr
```

Train And Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_arr, Y_arr, test_size = 0.01, shuffle = True, random_state=1)

print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

Create The Model

```python
def get_model():

  model = Sequential([
      Dense(10, input_shape = (5,), activation = 'relu'),
      Dense(20, activation = 'relu'),
      Dense(5, activation = 'relu'),
      Dense(1)
  ])

  model.compile(
      loss='mse',
      optimizer='adadelta'
  )

  return model
```

```python
model = get_model()
model.summary()
```

Model Training

```python
model = get_model()

# this prediction is before training the model
preds_on_untrained = model.predict(X_test)
```

```python
# Train model and store in the object history
history = model.fit(
    X_train, y_train,
    validation_data = (X_test, y_test),
    epochs = 1000
)
```

PREDICTIONS


```python
# Make predictions on the trained model
preds_on_trained = model.predict(X_test)
```
